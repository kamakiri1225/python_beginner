```python
import numpy as np
import pandas as pd
```

```python
data1 = np.random.randint(1, 10, 30).reshape(10,3)
data2 = np.random.randint(1, 10, 30).reshape(10,3)


df1 = pd.DataFrame(data1, columns=['a','b','c'])
df2 = pd.DataFrame(data2, columns=['a','b','c'])
```

```python
df1
```

```python
df2
```

```python
df1['a']
```

```python
df1[['a','b']]
```

```python
df1[['a']]
```

```python
df1.loc[5]
```

```python
pd.concat([df1, df2], axis=0)  #axis=0は行方向（下方向）に結合する
```

```python
pd.concat([df1, df2], axis=1) #axis=1は列方向（横方向）に結合する
```

```python
# 空のリストを用意
 
 
list_dct = []
for i in range(10):
  e = i+0
  f = i+1
  g = i+2
  h = i+3
  # 辞書型を作る
  dct = {
    'e':e,
    'f':f,
    'g':g,
    'h':h
  }
  list_dct.append(dct)
 
df1 = pd.DataFrame(list_dct)
df1
```

```python
# リスト型の中に辞書型を設定してデータフレームを作成
# 空のリストを用意
 
 
list_dct = []
for i in range(10):
  d = i-2
  c = i-1
  e = i+3
  f = i+10
  g = i+20
  h = i+30
 
  # 辞書型を作る
  dct = {
    'c':c,
    'd':d,
    'e':e,
    'f':f,
    'g':g,
    'h':h
  }
  list_dct.append(dct)
 
df2 = pd.DataFrame(list_dct)
df2
```

```python
df_merge = pd.merge(df1, df2, on='e', how='left')
df_merge
```

# グルーピングして統計量を見る

```python
df_merge['common'] = ['男性', '女性', '男性', '女性', '男性', '女性', '女性', '男性', '女性', '男性']
```

```python
df_merge
```

```python
# グルーピングの最大値を計算
df_merge.groupby('common').max()
```

```python
# グルーピングの最小値を計算
df_merge.groupby('common').min()
```

```python
# グルーピングの平均値を計算
df_merge.groupby('common').mean()
```

```python
# グルーピングの標準偏差を計算
df_merge.groupby('common').std()
```
